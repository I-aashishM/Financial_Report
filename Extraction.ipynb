```python
import pandas as pd #import library
```

```python
df = pd.read_excel('cik_list.xlsx') #reading excel files
df.head()
```

```python
df['SECFNAME'] = 'https://www.sec.gov/Archives/' + df['SECFNAME']  # concatenate the link with SECFNAME
```

```python
df.head()
```

Creating text files from SECFNAME (links) in cik_list.xlsx

```python
import requests
for i in range(len(df)):
    request = requests.get(df['SECFNAME'][i])
    text = request.text
    file = open('cik_list_text_files/'+ 'file'+str(i)+'.txt', 'w', encoding='utf-8')
    file.write(text)
    file.close()
```

Checking three sections in the text file

```python
from bs4 import BeautifulSoup
text_lists=[]
for i in range(len(df)):
    count = ''
    file = open('cik_list_text_files/'+'file'+str(i)+'.txt', 'r', encoding = 'utf-8')
    text = file.read()
    file.close()
    soup=BeautifulSoup(text,'html.parser')
    text_line = soup.get_text().lower()
    if "management's discussion and analysis" in text_line:
        count+='1'
    elif "quantitative and qualitative disclosures about market risk" in text_line:
        count+='2'
    elif "risk factors" in text_line:
        count+='3'
    text_lists.append(count)
#     print(text_lists)
```

There are 7 stop word files in stop_word_files folder. Converting text into lowercase

```python
from itertools import chain
from glob import glob
import os
path = r'Assignment_solution\stop_word_files'
for filename in os.listdir(path):
    if filename.endswith(".txt"):     
        file = open(os.path.join(path, filename), 'r')
        lines = [line.lower() for line in file]
    
    with open(os.path.join(path, filename), 'w') as out:
         out.writelines(lines)
```

Creating text file of section in cik_list_section_text folder from text file present in cik_list_text_files folder.......
 (number of file)_text1.text = it contain (management's discussion and analysis) content........
 (number of file)_text2.text = it contain (quantitative and qualitative disclosures about market risk) content..........
 (number of file)_text3.text = it contain (risk factors) content

```python
import re
for i in range(len(df)):
    f_obj=open('cik_list_text_files/'+'file'+str(i)+'.txt', 'r', encoding = 'utf-8')
    text=f_obj.read().lower()
    f_obj.close()
    soup=BeautifulSoup(text,'html.parser')
    if '1' in text_lists[i]:
        text1=''
        for elem in soup(text=re.compile("management's discussion and analysis")):
            text1=text1+' '+str(elem.parent)
            f_obj1=open('cik_list_section_text/'+str(i)+'_text1.txt','w',encoding='utf-8')
            f_obj1.write(text1)
            f_obj1.close()
    if '2' in text_lists[i]:
        text2=''
        for elem in soup(text=re.compile("quantitative and qualitative disclosures about market risk")):
            text2=text2+' '+str(elem.parent)
            f_obj1=open('cik_list_section_text/'+str(i)+'_text2.txt','w',encoding='utf-8')
            f_obj1.write(text2)
            f_obj1.close()
    if '3' in text_lists[i]:
        text3=''
        for elem in soup(text=re.compile("risk factors")):
            text3=text3+' '+str(elem.parent)
            f_obj1=open('cik_list_section_text/'+str(i)+'_text3.txt','w',encoding='utf-8')
            f_obj1.write(text3)
            f_obj1.close()
```
