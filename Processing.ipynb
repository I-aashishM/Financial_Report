## Calculating number of syllable

```python
def syllables(word):
    count = 0
    vowels = ['a','e','i','o','u']
    word = word.lower()
    
    if word[0] in vowels:
        count +=1
    for index in range(len(word)):
        if word[index] in vowels and word[index-1] not in vowels:
            count +=1
    if word.endswith('es') or word.endswith('ed'):
        count = 0
        
    return count
```
